```python
import pandas as pd
import re
```

```python
df = pd.read_csv('tweetData/Tweet Data/new_combined_tweets.csv')
df
```

# Normalization

```python
df['prompt_text'] = df['prompt_text'].str.lower()
df['response_text'] = df['response_text'].str.lower()
df
```

# Emojis, Hashtags, and Unicode characters.

```python
def clean_text(text):
    # Remove usernames starting with '@'
    text = re.sub(r'@\w+', '', text)
    # Remove specific patterns like 'Ax03FBB7BD5'
    text = re.sub(r'\bAx[A-Fa-f0-9]+\b', '', text)
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)
    # Remove emojis and non-ASCII characters
    text = text.encode('ascii', 'ignore').decode('ascii')
    # Remove special characters except question marks and dollar signs
    text = re.sub(r'[^\w\s?]', '', text)
    # Remove newline characters
    text = text.replace('\n', ' ')
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply the cleaning function to both columns
df['prompt_text'] = df['prompt_text'].apply(clean_text)
df['response_text'] = df['response_text'].apply(clean_text)
```
