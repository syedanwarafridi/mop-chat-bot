# Load Data

```python
import json
import pandas as pd

# Load JSON file
with open("tweetData/Tweet Data/Tweet_6k_27Feb.json", "r", encoding="utf-8") as file:
    data = json.load(file)  # Load JSON as a Python dictionary

rows = []
for key, entry in data.items():  # Iterate over dictionary values
    prompt = entry["prompt"]
    for response in entry.get("responses", []):  # Ensure 'responses' exists
        rows.append({
            "prompt_author": prompt["author"],
            "prompt_text": prompt["text"],
            "prompt_timestamp": prompt["timestamp"],
            "response_author": response["author"],
            "response_text": response["text"],
            "response_timestamp": response["timestamp"],
        })

# Convert to DataFrame
df = pd.DataFrame(rows)

df
```

```python
import os

directory = "tweetData/Tweet Data/"

df_list = []

for filename in os.listdir(directory):
    if filename.endswith(".json"):  
        filepath = os.path.join(directory, filename)

        with open(filepath, "r", encoding="utf-8") as file:
            data = json.load(file)

        if isinstance(data, list):
            rows = []
            for entry in data:
                prompt = entry["prompt"]
                for response in entry.get("responses", []):
                    rows.append({
                        "prompt_author": prompt["author"],
                        "prompt_text": prompt["text"],
                        "prompt_timestamp": prompt["timestamp"],
                        "response_author": response["author"],
                        "response_text": response["text"],
                        "response_timestamp": response["timestamp"],
                    })

            if rows: 
                df_list.append(pd.DataFrame(rows))

if df_list:
    final_df = pd.concat(df_list, ignore_index=True)
else:
    final_df = pd.DataFrame(columns=["prompt_author", "prompt_text", "prompt_timestamp", 
                                     "response_author", "response_text", "response_timestamp"])


final_df.to_csv("tweetData/Tweet Data/new_combined_tweets.csv", index=False, encoding="utf-8")
```

```python
def get_duplicates(df):
    df_reset = df.reset_index()  # Reset index to include it as a column
    duplicate_data = []

    for column in df_reset.columns:
        duplicates = df_reset[column].duplicated(keep=False)
        if duplicates.any():
            duplicate_values = df_reset.loc[duplicates, column].unique()
            df_duplicates = pd.DataFrame({"column_name": column, "duplicates": duplicate_values})
            duplicate_data.append(df_duplicates)

    if duplicate_data:
        return pd.concat(duplicate_data, ignore_index=True)
    else:
        return pd.DataFrame(columns=["column_name", "duplicates"])
```

```python
duplicates = get_duplicates(data)
duplicates.head(91)
```

```python
def remove_duplicates(df):
    # Keep only the first occurrence of each duplicate value within the same column
    df_cleaned = df.drop_duplicates(subset=["prompt_text", "response_text"], keep="first").reset_index(drop=True)
    return df_cleaned

# Usage:
df_cleaned = remove_duplicates(data)
```
